# **Assessment-if-else**

```python
# Assessment 1: Discount Eligibility
# Scenario:
# A store gives discounts based on the amount spent.
#Task:
# Write a Python program that:
# Takes the purchase amount as input.
# If the amount is greater than or equal to 100, the customer gets a 10% discount.
# Otherwise, print “No discount applied.”
# Example Input/Output:
# Enter purchase amount: 120
# You get a 10% discount. Final price: 108.0

purchase_amount=int(input("Enter Purchase amount:"))

if purchase_amount>=100:
  print("Final price:",purchase_amount*0.9)
else:
  print("No discount Applied")
```

```python
# Assessment 2: Temperature Checker
# Scenario:
# You’re creating a basic weather assistant.
# Task:
# Write a Python program that:
# Takes the current temperature as input.
# If the temperature is below 0, print “Freezing cold”.
# If between 0 and 20, print “Cold weather”.
# Otherwise, print “Warm day”.
# Example:
# Enter temperature: 15
# Cold weather

temp_input=int(input("Current Temperature:"))

if temp_input<0:
  print("Freezing cold")
elif (temp_input>0 & temp_input<20):
  print("Cold weather")
else:
  print("Warm Day")
```

```python
# Assessment 3: Login System
# Scenario:
# Simulate a simple login check.
# Task:
# Write a program that:
# Asks for a username and password.
# If the username is "admin" and password is "1234", print “Login successful”.
# If username is "admin" but password is wrong, print “Incorrect password”.
# Otherwise, print “User not found”.
# Hint: Use nested if statements.

username=input("Enter username:")
password=input("Enter Password:")

if username=="admin" and password=="1234":
  print("Login successful")
elif username=="admin" and password!="1234":
  print("Incorrect password")
else:
  print("User not found")
```

```python
# Assessment 4: Student Grade Evaluator
# Scenario:
# You’re building a grade evaluator.
# Task:
# Ask for a student’s marks (0–100).
# If marks ≥ 90 → Grade A
# Else if marks ≥ 75 → Grade B
# Else if marks ≥ 50 → Grade C
# Else → Fail
# Add a nested if: If the student got Grade A and marks == 100, print “Perfect Score!”.
# Example:
# Enter marks: 100
# Grade: A
# Perfect Score!

marks=int(input("Enter Students marks:"))

if marks>= 90:
  print("Grade A")
  if marks==100:
    print("Perfect Score!")
elif marks>=75:
  print("Grade B")
elif marks>=50:
  print("Grade C")
else:
  print("Fail")
```

```python
# Assessment 5: Loan Eligibility Checker
# Scenario:
# You work for a bank’s software system.
# Task:
# Write a Python program that:
# Asks the user for:
# Age
# Annual income
# Credit score
# Use nested if-else logic to decide:
# If age < 18 → “Not eligible: underage”
# Else:
# If income < 25000 → “Not eligible: low income”
# Else:
# If credit score < 600 → “Not eligible: poor credit score”
# Else → “Eligible for loan”
# Example:
# Enter age: 30
# Enter income: 40000
# Enter credit score: 700
# Eligible for loan

age=int(input("Enter age:"))
income=int(input("Enter Annual Income:"))
score=int(input("Enter Credit Score:"))

if age<18:
  print("Not eligible: underage")
else:
  if income < 25000:
    print("Not eligible: low income")
  else:
    if score < 600:
      print("Not eligible: poor credit score")
    else:
      print("Eligible for loan")
```
