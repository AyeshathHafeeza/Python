```python
# Assessment 1: Daily Task Reminder (for loop)
# Scenario:
# You’re helping a productivity app display a user’s daily tasks.
# Task:
# Write a Python program that:
# Stores a list of tasks:
# tasks = ["Check emails", "Attend meeting", "Complete report", "Call client"]
# Use a for loop to print each task with a number (starting from 1).
# Expected Output:
# 1. Check emails
# 2. Attend meeting
# 3. Complete report
# 4. Call client

tasks = ["Check emails", "Attend meeting", "Complete report", "Call client"]
for num in range(0,4):
    print(f"{num+1}.{tasks[num]}")
```

```python
# Assessment 2: Countdown Timer (while loop)
# Scenario:
# You’re simulating a simple countdown timer before starting an event.
# Task:
# Write a Python program that:
# Takes a number from the user (e.g., 5)
# Uses a while loop to count down to 1
# Prints “GO!” after the countdown ends.
# Example Input/Output:
# Enter countdown start: 5
# 5
# 4
# 3
# 2
# 1
# GO!

num=int(input("Enter number:"))
while(num>=0):
  print(num)
  num-=1

print("Go!")

```

```python
# Assessment 3: Sum of Even Numbers (for loop)
# Scenario:
# You’re developing a feature for a budgeting app that sums only even transaction IDs.
# Task:
# Write a Python program that:
# Takes a list of numbers (e.g. [12, 5, 8, 3, 10, 15])
# Uses a for loop and if statement to find and sum only even numbers.
# Prints the total sum.
# Expected Output:
# Total of even numbers: 30
# Hint: Use the modulus operator % to check even numbers.

lst=[12, 5, 8, 3, 10, 15]
sum=0
for i in lst:
  if i%2==0:
    sum=sum+i
print(sum)
```

```python
# Assessment 4: Password Retry System (while loop)
# Scenario:
# You’re implementing a login retry system.
# Task:
# Write a program that:
# Has a preset password: "python123"
# Allows the user 3 attempts to enter the correct password.
# If the user enters the correct password, print "Access Granted" and stop the loop.
# If they fail all attempts, print "Account Locked".
# Example Output:
# Enter password: test
# Wrong password.
# Enter password: hello
# Wrong password.
# Enter password: python123
# Access Granted

for i in range(0,3):
  n=input("Enter the password:")
  if n=="python123":
    print("Access Granted")
    break
  else:
    print("Wrong Password")
else:
  print("Account Locked")
```
