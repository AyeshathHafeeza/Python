```python
# 6. List Comprehension & Filtering
# Task:
# Given the list nums = [1, 2, 3, 4, 5, 6, 7, 8, 9],
# Use list comprehension to create a new list of only even numbers squared.
# Output should look like [4, 16, 36, 64].
# Topics: List Comprehension

list= [1, 2, 3, 4, 5, 6, 7, 8, 9]
new_list=[x**2 for x in list if x%2==0]
print(new_list)
```

```python
# 8. Working with Tuples and Sets Together
# Task:
# Create a tuple colors = ("red", "green", "blue", "yellow")
# Convert it into a set
# Add a new color "orange" to the set
# Explain why you cannot add to the tuple directly
# Topics: Tuples (immutability), Sets (mutability)

colors= ("red", "green", "blue", "yellow")
new_set=set(colors)
print(new_set)

new_set.add("orange")
print(new_set)

#tuple is immutable
```

```python
# 9. Dictionary Operations with Loop
# Task:
# Create a dictionary of 3 employees with their salaries:
# salaries = {"Ali": 50000, "Sara": 60000, "John": 55000}
# Use a for loop to print each employee’s name and salary increased by 10%.
# Topics: Dictionaries, Loops, Operators

employee={"Ali": 50000, "Sara": 60000, "John": 55000}

for name, salary in employee.items():
    new_salary = salary * 1.10
    print(f"{name}: {new_salary}")
```

```python
# 10. Combined Data Structures Mini Task
# Task:
# Create a list of dictionaries to represent students:
# students = [
#     {"name": "Ali", "marks": 85},
#     {"name": "Sara", "marks": 92},
#     {"name": "John", "marks": 78}
# ]
# Use:
# A loop to print all names with marks > 80
# A list comprehension to create a list of only the marks
# A set to remove duplicate marks if any
# Topics: Lists, Dictionaries, Loops, List Com

students = [
     {"name": "Ali", "marks": 85},
    {"name": "Sara", "marks": 92},
    {"name": "John", "marks": 78}
]
for student in students:
    if student["marks"] > 80:
        print(student["name"])

marks = [student["marks"] for student in students]
print(marks)

unique_marks = set(marks)
print(unique_marks)
```
