```python
#1.Write a Python function normalize_text(text) that:
#removes leading and trailing spaces.
#Converts the string to lowercase.
#Replaces all tabs (\t) with a single space.
#Replaces multiple spaces with a single space.
#Returns the cleaned string.
import re
def normalize_text(text):
  text=text.strip()
  text=text.lower()
  text=text.replace("\\t"," ")
  text = re.sub(r'\s+', ' ', text)
  return text

raw_text=input("Enter string:")
cleaned=normalize_text(raw_text)
print(cleaned)
```

```python
#2.Write a function word_frequency(sentence) that counts how many times each word appears in a given sentence (case-insensitive) and returns a dictionary.
#Example:
#Input:  "Python is fun, and python is powerful."
#Output: {'python': 2, 'is': 2, 'fun,': 1, 'and': 1, 'powerful.': 1}
input="Python is fun, and python is powerful"
def word_frequency(sentence):
  sentence=sentence.lower()
  words=sentence.split()
  freq={}

  for word in words:
    if word in freq:
      freq[word]+=1
    else:
      freq[word]=1
  return freq

text="Python is fun, and python is powerful"
result = word_frequency(text)
print(result)
```
